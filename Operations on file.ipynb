## Renaming column

```python
#using dictionary substitution 
data=data.rename(columns={'car name':'car_name','model year':'model_year'})
```

```python
data.head()
```

```python
data.columns
```

```python
#using list 
#only changing hoursepower to hp, but we need to list all the column names
data.columns=['mpg', 'cylinders', 'displacement', 'hp', 'weight',
       'acceleration', 'model_year', 'origin', 'car_name']
```

```python
data.head()
```

## Deleting Columns

```python
#drop will allow to delete multiple columns
#need to assign..no change in source
data1=data.drop(columns=['origin'])
```
