# histogram

```python
data = pd.read_csv('auto-mpg.csv')
data.head()
```

```python
data.columns
```

```python
plt.figure()
plt.hist(data['mpg'])
plt.title('Distribution of MPG')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'])
plt.title('Distribution of MPG')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'])
plt.title('Distribution of MPG',fontsize=16)
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red')
plt.title('Distribution of MPG',fontsize=16)
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red')
plt.title('Distribution of MPG',fontsize=16)
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red',rwidth=0.9)
plt.title('Distribution of MPG',fontsize=16)
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red',rwidth=0.9,density=True)
plt.title('Distribution of MPG',fontsize=16)
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red',rwidth=0.9,density=True,cumulative=True)
plt.title('Distribution of MPG',fontsize=16)
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
#changing datatype
data['horsepower']=data['horsepower'].astype(float)
```

```python
#delete those rows using drop function..it doesnt save the change on the file so we assign it to a new file
data1=data.drop([32, 126, 330, 336, 354, 374])
```

# Scatter plot

```python
plt.figure(figsize=(8,4))
plt.scatter(data1['horsepower'],data1['mpg'],s=10)
plt.title('Plot of HP vs MPG',fontsize=16)
plt.xlabel('HP')
plt.ylabel('MPG')
```

```python
plt.figure(figsize=(8,4))
plt.scatter(data1['horsepower'],data1['mpg'],s=10,c='violet')
plt.title('Plot of HP vs MPG',fontsize=16)
plt.xlabel('HP')
plt.ylabel('MPG')
```

```python
plt.figure(figsize=(8,4))
plt.scatter(data1['horsepower'],data1['mpg'],s=20,c='red',marker='^')
plt.title('Plot of HP vs MPG',fontsize=16)
plt.xlabel('HP')
plt.ylabel('MPG')
```

```python
plt.figure(figsize=(8,4))
plt.scatter(data1['displacement'],data1['mpg'],s=15,c='red',marker='^')
plt.title('Plot of Disp vs MPG',fontsize=16)
plt.xlabel('Disp')
plt.ylabel('MPG')
```

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.scatter(data1['horsepower'],data1['mpg'],s=12,c='red',marker='*')
plt.title('Plot of HP vs MPG',fontsize=10)
plt.xlabel('HP',fontsize=10)
plt.ylabel('MPG',fontsize=10)

plt.subplot(2,1,2)
plt.scatter(data1['displacement'],data1['mpg'],s=12,c='red',marker='*')
plt.title('Plot of Disp vs MPG',fontsize=10)
plt.xlabel('Disp',fontsize=10)
plt.ylabel('MPG',fontsize=10)

plt.tight_layout()
```
